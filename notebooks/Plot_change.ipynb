```python
import xarray as xr
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import cmocean as cmo
```

```python
d0 = 20
d1 = 102
dpy = 2

ds0 = xr.open_dataset(f'../results/CD_1.0_tanh_Tdeep0.5_ztcl-500_{d0:03.0f}.nc')
ds1 = xr.open_dataset(f'../results/CD_1.0_tanh_Tdeep0.5_ztcl-500_{d1:03.0f}.nc')

Nyears = (d1-d0)/dpy

fig,ax = plt.subplots(2,3,figsize=(10,7))

for m in [0,1,2]:
    for n in [0,1]:
        ax[n,m].pcolormesh(ds0.mask,cmap='cmo.ice',vmin=-3,vmax=3)
        
im = ax[0,0].pcolormesh(np.where(ds0.tmask==1,ds0.zb,np.nan),vmin=-1500,vmax=0,cmap='gist_stern_r')
plt.colorbar(im,ax=ax[0,0],orientation='horizontal')
ax[0,0].text(.1,.9,f'{np.nanmean(np.where(ds0.tmask==1,ds0.zb,np.nan)):+.0f} m',transform=ax[0,0].transAxes)

im = ax[1,0].pcolormesh(np.where(ds0.tmask==1,ds0.melt,np.nan),norm=mpl.colors.LogNorm(vmin=1, vmax=100),cmap='inferno')
plt.colorbar(im,ax=ax[1,0],orientation='horizontal')
ax[1,0].text(.1,.9,f'{np.nanmean(np.where(ds0.tmask==1,ds0.melt,np.nan)):+.1f} m/yr',transform=ax[1,0].transAxes)



im = ax[0,1].pcolormesh(np.where(ds1.tmask==1,ds1.zb,np.nan),vmin=-1500,vmax=0,cmap='gist_stern_r')
plt.colorbar(im,ax=ax[0,1],orientation='horizontal')
ax[0,1].text(.1,.9,f'{np.nanmean(np.where(ds1.tmask==1,ds1.zb,np.nan)):+.0f} m',transform=ax[0,1].transAxes)

im = ax[0,2].pcolormesh(np.where(ds0.tmask==1,ds1.zb-ds0.zb,np.nan),vmin=-100,vmax=100,cmap='cmo.balance')
plt.colorbar(im,ax=ax[0,2],orientation='horizontal')
ax[0,2].text(.1,.9,f'{np.nanmean(np.where(ds0.tmask==1,ds1.zb-ds0.zb,np.nan)):+.0f} m',transform=ax[0,2].transAxes)




im = ax[1,1].pcolormesh(np.where(ds0.tmask==1,ds1.melt,np.nan),norm=mpl.colors.LogNorm(vmin=1, vmax=100),cmap='inferno')
plt.colorbar(im,ax=ax[1,1],orientation='horizontal')
ax[1,1].text(.1,.9,f'{np.nanmean(np.where(ds0.tmask==1,ds1.melt,np.nan)):+.1f} m/yr',transform=ax[1,1].transAxes)

im = ax[1,2].pcolormesh(np.where(ds0.tmask==1,(ds1.melt-ds0.melt),np.nan),vmin=-10,vmax=10,cmap='cmo.balance')
plt.colorbar(im,ax=ax[1,2],orientation='horizontal')
ax[1,2].text(.1,.9,f'{np.nanmean(np.where(ds0.tmask==1,(ds1.melt-ds0.melt),np.nan)):+.1f} m/yr',transform=ax[1,2].transAxes)

ax[0,0].set_title('Draft year 0')
ax[0,1].set_title(f'Draft year {Nyears}')
ax[0,2].set_title('Difference (red: thinning ice)')

ax[1,0].set_title('Melt year 0')
ax[1,1].set_title(f'Melt year {Nyears}')
ax[1,2].set_title('Difference (red: increased melt)')

for n in [0,1,2]:
    for m in [0,1]:
        ax[m,n].set_xticks([])
        ax[m,n].set_yticks([])
ds0.close()
ds1.close()
```
